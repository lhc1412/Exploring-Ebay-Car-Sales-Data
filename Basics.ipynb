# Exploring Ebay Car Sales Data

Th aim of this project is to clean the data and analyze the included used car listings. The dataset of used cars is from eBay Kleinanzeigen, a classifieds section of the German eBay website. The aim of this project is to clean the data and analyze the included used car listings.

### Summary of results

After analyzing the data, the key findings are following:

- Amoung the top 20 brands, the brand with the highest mean price is mini, following by audi and mercedes_benz.
- The brand with the lowest mean price is renault. Following by fiat and Opel.
- No obvious relationship between the mean price and the mean mileage of the cars can be found
- If we only consider the mean price and mean mileage, smart, hyundai and skoda are the best options as theyhave relatively low mean mileage but with also low mean price.

For more details, please refer to the the full analysis below.

The column names were converted from camelcase to snakecase and some of the column names were reworded based on the data dictionary to be more descriptive.

### Converting numeric data which stored as text to numeric data

In the following part, the text data stored in "price" and "odometer" will be converted to numeric data.

```python
#convert price to numeric
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(int)

#confirm results
autos['price'].describe()
```

```python
#convert odometer to numeric
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','').astype(int)
autos.rename({'odometer': 'odometer_km'}, axis=1, inplace=True)

#confirm results
autos['odometer_km'].describe()
```

## Data analysis

### Exploring variations across different car brands

The average value of the top 20 brands will be explored.

```python
# show top 20 car brands
top_20_brands = autos['brand'].value_counts().head(20).index
top_20_brands
```

```python
# find out mean price for top 20 brands
top_20_brands_mean_price = {}
for brand in top_20_brands:
    top_20_brands_mean_price[brand] = int(round(autos.loc[autos['brand'] == brand, "price"].mean()))
    
top_20_brands_mean_price
```

The brand with the highest mean price is mini (USD10654), following by "sonstige_autos", which is not a brand, it means "other cars". The following brands are audi (USD9358) and mercedes_benz (USD8551). The brand with the lowest mean price is renault, the mean price is USD2469 only.

### Comparing average mileage and mean price

Aggregation will be used to understand the average mileage for those cars and if there's any visible link with mean price. 

```python
# assign mean price to new list for comparison
brand_series = pd.Series(top_20_brands_mean_price)
brand_compare = pd.DataFrame(brand_series, columns=['mean_price'])
brand_compare
```

```python
# find out mean mileage 
top_20_brands_mean_mileage = {}
for brand in top_20_brands:
    top_20_brands_mean_mileage[brand] = int(round(autos.loc[autos['brand'] == brand, "odometer_km"].mean()))
    
top_20_brands_mean_mileage
```

```python
# put mean mileage into comparing list
mile_series = pd.Series(top_20_brands_mean_mileage)
brand_compare['mean_mileage'] = mile_series
brand_compare
```

We can say that there is no obvious relationship between the mean price and the mean mileage of the cars. mini has the highest mean price and lowest mean mileage, sonstige_autos (other cars) are has the second highest mean price and second lowest mean mileage. Besides this two, smart, hyundai and skoda have relatively low mean mileage but with also low mean price. For other ar brands in high price range (audi, bmw and mercedes benz), they have driven in average more than 129000km.
